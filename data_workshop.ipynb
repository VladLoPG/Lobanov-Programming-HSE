#### Лингвистический анализ

1. Небольшая выборка для аннотирования данных

```python
print(len(filtered_data))
```

```python
import random

filtered_data = [i for i in cleaned_data if len(i) > 0] # если длина текста из cleaned_data > 0, добавляем его в список filtered_data
sample_size = max(1, int(len(filtered_data) * 0.1))  # задаем размер выборки: 10%
sample_data = random.sample(filtered_data, sample_size) # семплирование
sample_text = ' '.join(sample_data) # объедините все в одну строку
print(sample_text)
```

2. Лемматизация

```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp(sample_text)
lemmas = [token.lemma_ for token in doc]

for i in range(30):
  print(lemmas[i], doc[i]) # выведите пары (token, lemma) в диапазоне 30,
# т.е. для первых 30 элементов списка lemmas и для первых 30 элементов doc
```

3. Частеречная разметка

```python
pos_annotation = [(token.text, token.pos_) for token in doc]

pos_annotation[:30] # выведите пары (token, pos) в диапазоне 30
```

4. Синтаксическая разметка

```python
from spacy import displacy

doc_ = nlp('This is a sample sentence.')

syntax_annotation = [(token.text, token.dep_, token.head.text) for token in doc_]
displacy.render(doc_, style="dep", jupyter=True, options={'distance': 150})
```

5. Разметка Named Entity Recognition

```python
ner_annotation = [(ent.text, ent.label_) for ent in doc.ents]
displacy.render(doc, style="ent", jupyter=True)
```
