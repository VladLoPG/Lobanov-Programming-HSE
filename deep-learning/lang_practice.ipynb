```python
import getpass
import os

if not os.environ.get("GROQ_API_KEY"):
  os.environ["GROQ_API_KEY"] = getpass.getpass("Enter API key for Groq: ")

from langchain.chat_models import init_chat_model

model = init_chat_model("meta-llama/llama-4-maverick-17b-128e-instruct", model_provider="groq", temperature=0.5)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage(content="Translate the following from English into Russian"),
    HumanMessage(content="Hi! I am dying to see you. It's been such a long time. Are you even the same person?"),
]

print(model.invoke(messages).content)
```
